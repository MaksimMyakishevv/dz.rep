### Применяем токенизаторы

*WhitespaceTokenizer*

```python
start_time = timeit.default_timer()

whitespace = nltk.tokenize.WhitespaceTokenizer()
print(whitespace.tokenize(text))

print(timeit.default_timer() - start_time)
```

*WordPunctTokenizer*

```python
start_time = timeit.default_timer()

wordpunct = nltk.tokenize.WordPunctTokenizer()
print(wordpunct.tokenize(text))

print(timeit.default_timer() - start_time)
```

*NLTKWordTokenizer*

```python
start_time = timeit.default_timer()

word_tok = nltk.tokenize.NLTKWordTokenizer()
print(word_tok.tokenize(text))

print(timeit.default_timer() - start_time)
```

```python
tokenized_wordpunct = nltk.tokenize.WordPunctTokenizer().tokenize(text)
tokenized_space = nltk.tokenize.SpaceTokenizer().tokenize(text)
tokenized_mwe = nltk.tokenize.MWETokenizer().tokenize(text)
tokenized_line = nltk.tokenize.LineTokenizer().tokenize(text)
tokenized_punktsentence = nltk.tokenize.PunktSentenceTokenizer().tokenize(text)
tokenized_sexpr = nltk.tokenize.SExprTokenizer().tokenize(text)
tokenized_syllable = nltk.tokenize.SyllableTokenizer().tokenize(text)
tokenized_tab = nltk.tokenize.TabTokenizer().tokenize(text)
tokenized_toktok = nltk.tokenize.ToktokTokenizer().tokenize(text)
tokenized_treebankword = nltk.tokenize.TreebankWordTokenizer().tokenize(text)
tokenized_nltkword = nltk.tokenize.NLTKWordTokenizer().tokenize(text)
tokenized_tweet = nltk.tokenize.TweetTokenizer().tokenize(text)
tokenized_regexp = nltk.tokenize.RegexpTokenizer(r"\w+").tokenize(text)
tokenized_whitespace = nltk.tokenize.WhitespaceTokenizer().tokenize(text)
tokenized_wordtokenize = nltk.tokenize.word_tokenize(text)
tokenized_senttokenize = nltk.tokenize.sent_tokenize(text)
tokenized_blankline = nltk.tokenize.BlanklineTokenizer().tokenize(text)
tokenized_legalitysyllable = nltk.tokenize.LegalitySyllableTokenizer(text).tokenize(text)
#tokenized_texttiling = nltk.tokenize.TextTilingTokenizer().tokenize(text)
```

```python
class Token:
    @staticmethod
    def token(text):
        for punkt in ('.', ',', ';', ':', '-','!','?'):
            tep = text.replace(punkt, " ")
        return tep.split()
token_custom = Token.token(text)
token_custom
```

```python
import spacy
from spacy.language import Language
spacy.cli.download("en_core_web_sm")
```

```python
@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text == "...":
            doc[token.i + 1].is_sent_start = True
    return doc

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("set_custom_boundaries", before="parser")
ellipsis_doc = nlp(text)
ellipsis_sentences = list(ellipsis_doc.sents)
for sentence in ellipsis_sentences:
    print(sentence)
    print("\n===SENTENCE END===\n")
```
