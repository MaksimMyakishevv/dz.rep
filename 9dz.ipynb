```python


def vector_mul(vec1, vec2):
        res = 0
        for i in range(len(vec2[0])):
            res += vec1[0][i] * vec2[0][i]
        return res
def norm(vec):
    print(vec.matrix)
    result = vec.matrix[0][0]**2 + vec.matrix[0][1]**2 + vec.matrix[0][2]**2 + vec.matrix[0][3]**2
    return result**0.5

class Matrix:
    def __init__(self,matrix):
        self.matrix = matrix

    def __add__ (self, other):
        res = [] 
        if isinstance(other,int) or isinstance(other,float): 
           for line in range(len(self.matrix)):
                new = [] 
                for column in range(len(self.matrix[0])):   
                    new.append (self.matrix[line][column] + other)  
                res.append(new) 
        else:
            for line in range(len(self.matrix)):
                new = [] 
                for column in range(len(self.matrix[0])):
                    new.append (self.matrix[line][column] + other.matrix[line][column] )  
                res.append(new)
        return Matrix(res)
    
    def __sub__ (self, other):
        res = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):   
                    new.append (self.matrix[line][column] - other)  
                res.append(new)
        else:
            for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):  
                    new.append (self.matrix[line][column] - other.matrix[line][column] )  
                res.append(new)
        return Matrix(res)
                
    def __truediv__ (self, other):
        res = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):   
                    new.append (self.matrix[line][column] / other)  
                res.append(new)
        else:
            for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):  
                    new.append (self.matrix[line][column] / other.matrix[line][column] )  
                res.append(new)
        return Matrix(res)
    
    def __matmul__(self,other):
        res = [] 
        for line in range(len(self.matrix)):
            new = [] 
            for column in range(len(other.matrix[0])):
                skolar = 0  
                for i in range(len(self.matrix[0])):  
                    skolar += self.matrix[line][i] * other.matrix[i][column] 
                new.append(skolar)
        res.append(new)
    
        return Matrix(res)
    
    def __mul__(self,other):
        res = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):
                    new.append (self.matrix[line][column] * other)  
                res.append(new)
        else:
            for line in range(len(self.matrix)):
                new = []
                for column in range(len(self.matrix[0])):   
                    new.append (self.matrix[line][column] * other.matrix[line][column] )  
                res.append(new)
        return Matrix(res)  
    
    def T(self):
        res = [] 
        for column in range(len(self.matrix[0])): 
            new = []
            for line in range(len(self.matrix)): 
                new.append(self.matrix[line][column])
            res.append (new)
        return Matrix(res)
    
    def get_v(self):
        v1, v2, v3, v4 = [],[],[],[]
        for i in self.matrix:
            v1.append(i[0])
            v2.append(i[1])
            v3.append(i[2])
            v4.append(i[3])
        return [v1,], [v2,], [v3,], [v4,]


    def bas(self):
        v1 = self.get_v()[0]
        v2 = self.get_v()[1]
        v3 = self.get_v()[2]
        v4 = self.get_v()[3]

        f1 = v1

        f2 = (Matrix(v2) - (Matrix(f1)*(vector_mul(v2,f1) / vector_mul(f1,f1)))).matrix

        f3 = (Matrix(v3) - (Matrix(f1)*(vector_mul(v3,f1) / vector_mul(f1,f1))) -
                            (Matrix(f2)*(vector_mul(v3,f2) / vector_mul(f2,f2)))).matrix
        
        f4 = (Matrix(v4) - (Matrix(f1)*(vector_mul(v4,f1) / vector_mul(f1,f1))) -
                            (Matrix(f2)*(vector_mul(v4,f2) / vector_mul(f2,f2))) -
                            (Matrix(f3)*(vector_mul(v4,f3) / vector_mul(f3,f3)))).matrix
        return Matrix(f1), Matrix(f2), Matrix(f3), Matrix(f4)
    
    def ort(self):
        res = []
        for f in self.bas():
            res.append((f/norm(f)).matrix)
        return(res)
```
